```python
import numpy as np 
import torch 
from nnModel import Net
import wandb
```

```python
run = wandb.init(
    # Set the wandb entity where your project will be logged (generally your team name).
    entity="sanjana-learning",
    # Set the wandb project where this run will be logged.
    project="cat-or-dog",
    # Track hyperparameters and run metadata.
    config={
        "architecture": "CNN",
        "dataset": "https://www.kaggle.com/datasets/shaunthesheep/microsoft-catsvsdogs-dataset/data",
    },
)
```

```python
img_size = 50

net = Net()
net.load_state_dict(torch.load("/model/catdog_cnn_model_1.pth"))
net.eval()  # Set the model to evaluation mode

X_test_data = np.load("/model/catdog1_X_test.npy", allow_pickle=True)
y_test_data = np.load("/model/catdog1_y_test.npy", allow_pickle=True)

X_test = torch.Tensor([img for img in X_test_data])
y_test = torch.Tensor([dx for dx in y_test_data])
```

```python
with torch.no_grad():
    iteration_count = 0
    correct = 0
    total = 0  
    for i in range(len(X_test)):
        logits = net(X_test[i].view(1, 1, img_size, img_size))  # (1,2)
        pred = torch.argmax(logits, dim=1).item()               # 0 or 1
        true = torch.argmax(y_test[i]).item()                   # 0 or 1 (one-hot -> index)

        correct += (pred == true)
        total += 1

print(f"Accuracy: {correct}/{total} = {correct/total:.2%}")
accuracy = correct/total
run.log({"accuracy": accuracy})
```

```python
run.finish()
```
